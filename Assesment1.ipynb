```python
import json
def FilterOrder(orders):
    status=[]
    for order in orders:
        if order["amount"]>150:
            order["status"]="processed"
            status.append(order)
    return json.dumps(status,indent=4)
orders=[
    {"id":1,"amount":200},
    {"id":2,"amount":500},
    {"id":3,"amount":100},
    {"id":4,"amount":50}
]
print(FilterOrder(orders))
```

7.Create a logger decorator to log function name,number if args,number of kwargs and arguments passed to the function

```python
import functools
def logger_decorator(func):
    @functools.wraps(func)
    def wrapper(*args,**kwargs):
        print(f"Calling{func.__name__ }with args:{args} and kwargs:{kwargs}")
        result=func(*args,**kwargs)
        print(f"{func.__name__} returned{result}")
        return result
    return wrapper
@logger_decorator
def add(a,b):
    return a+b
result=add(2,3)
print(result)
def compute_sum(a,b,c=0):
    return a+b+c
result=compute_sum(1,2,c=3)
print(result)
```

```python
def Logs_input_output(func):
  def wrapper(*args,**kwargs):
    print("Inputs:args=",args,"kwargs=",kwargs)
    output=func(*args,**kwargs)
    print("Output:",output)
    return output
  return wrapper
@Logs_input_output
def Add(a,b):
  return a+b

a=int(input("Enter number to a:"))
b=int(input("Enter number to b:"))
result=Add(a,b)
print(result)
```

8.Create a generator function to read from large size file

```python
def LargeSizeFile(filepath):
  try:
    with open(filepath,'r') as file:
      for line in file:
        yield line
  except FileNotFoundError:
    print(f"File not found:{filepath}")
# filepath='Automation.txt'
filepath=input("Enter a file path:")
for line in LargeSizeFile(filepath):
  print(line)
```
