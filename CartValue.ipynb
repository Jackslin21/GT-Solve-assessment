```python
def CartValue(data):
    total = 0
    discount = 0

    for item in data:
       price=data[item] ['price']
       discount=data[item]['discount']

       total += price
       discount += (price * discount / 100)

    final_amount = total - discount
    print("Final amount to pay:",final_amount)
data={
        "shirts":{"price":100,"discount":20},
        "shoe":{"price":800,"discount":25},
        "clock":{"price":150,"discount":15},
        "watch":{"price":1500,"discount":5}
    }
CartValue(data)


```
