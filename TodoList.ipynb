```python
import requests

def Fetching():
    url = "https://jsonplaceholder.typicode.com/todos"
    response = requests.get(url)
    Todo = response.json()

    for todo in Todo[:5]:
        print(todo)

def AddTask():
  url = "https://jsonplaceholder.typicode.com/todos"
  newTask = {
      'userId': 10,
      'id': 201,
      'title' : 'Python TodoList',
      'completed': False

  }

  response = requests.post(url,json = newTask)
  print("Added Task : ",response.json())

def UpdateTask():
    url = "https://jsonplaceholder.typicode.com/todos/1"
    update = {
        'title' : 'Todo List in python',
        'completed' : True
    }
    response = requests.put(url,json = update)
    print("Updated : ",response.json())

def DeletedTask():
    url = "https://jsonplaceholder.typicode.com/todos/1"
    response = requests.delete(url)
    print("Deleted : ",response.status_code)


Fetching()
AddTask()
UpdateTask()
DeletedTask()
```
